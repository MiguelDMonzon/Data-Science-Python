Ejercicio 2.
* Crea dos arrays de 100 elementos con números aleatorios de distribuciones diferentes, pero en un rango de valores similar.
* Dibuja sus histogramas y compáralos.
* Busca en scypy.stats una función para hacer una comprobación (mediante un t-test) de si la media de los elementos en los dos arrays es diferente.

```python
from scipy.stats import expon

min_value = 0
max_value = 100

arr1 = np.random.uniform(min_value, max_value, 100)
arr2 = expon.rvs(min_value, max_value, 100)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.hist(arr1, normed=True, alpha=0.2, bins="auto")
plt.hist(arr2, normed=True, alpha=0.2, bins="auto")
```

```python
from scipy.stats import ttest_ind
print ttest_ind(arr2, arr1)
```

A continuación, queremos examinar el sexo y la edad de los implicados en incidentes.

Mostrad la distribución de los incidentes por sexo, e intentad obtener la media de la edad de los

Observad:
* Valores nulos o valores que parecen incorrectos.
* El tipo de los valores de la columna "Age". Si fuese necesario cambiarlos, buscar formas de convertir de valores de cadena a valores numéricos en pandas.

```python
cols = ["Sex ", "Age"]
by_sex = df[cols].dropna().groupby("Sex ")
by_sex.groups.keys()
```

Se aprecia que uno de los sexos en "M ", así que se entiende que quiere decir lo mismo que "M", por lo tanto se sustituye por este. También, en principio el sexo "lli" no se reconoce así que se cambiará por un valor nulo para después eliminarlo.

```python
by_sex = df[cols].replace("M ", "M").replace("lli", np.nan).dropna().groupby("Sex ")
by_sex.count()
```

```python
df["Age"].unique()
```

Como se puede observar, existen todo tipo de valores no numéricos dentro de esta columna. Puesto que se quiere calcular la media, se van a transformar todos a tipo entero y los que no se pueda, los transformará como NA (con errors = coerce).

```python
df["Age"] = pd.to_numeric(df["Age"], errors = "coerce")

df["Age"].dropna().mean()
#df["Age"].mean() también funciona
```
